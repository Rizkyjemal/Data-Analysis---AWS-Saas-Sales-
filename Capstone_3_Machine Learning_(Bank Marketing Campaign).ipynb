### **Confusion Matrix**

|                      | **Actual 1 (Deposit)** | **Actual 0 (No Deposit)** |
|----------------------|------------------------|---------------------------|
| **Pred 1 (Deposit)** | **TP – True Positive** : diprediksi deposit & benar deposit | **FP – False Positive** : diprediksi deposit tapi tidak deposit |
| **Pred 0 (No)**      | **FN – False Negative** : diprediksi tidak deposit padahal deposit | **TN – True Negative** : diprediksi tidak deposit & benar tidak deposit |

- **True Positive (TP):** Nasabah yang diprediksi akan membuka deposito dan benar-benar membuka deposito – *target ideal campaign*.
- **False Positive (FP):** Nasabah yang diprediksi akan membuka deposito, tetapi sebenarnya tidak – *biaya campaign terbuang*.
- **False Negative (FN):** Nasabah yang diprediksi tidak akan membuka deposito, tetapi sebenarnya membuka deposito – *kehilangan peluang konversi*.
- **True Negative (TN):** Nasabah yang diprediksi tidak akan membuka deposito dan memang tidak membuka deposito. – *hemat biaya*.

### **Evaluation Matrix**
- **Accuracy** = (TP + TN) / Total  
  -> Proporsi prediksi yang benar terhadap total data. 
- **Precision** = TP / (TP + FP)  
  -> Persentase nasabah yang benar-benar membuka deposito dari seluruh nasabah yang diprediksi akan membuka deposito, penting untuk *efisiensi biaya*.
- **Recall (Sensitivity)** = TP / (TP + FN)  
  -> Persentase nasabah yang benar-benar membuka deposito dan berhasil terprediksi oleh model, penting untuk *menghindari peluang hilang*.
- **F1 Score** = 2 × (Precision × Recall) / (Precision + Recall)  
  -> Keseimbangan Precision dan Recall.
- **ROC AUC / PR AUC**  
  -> Mengukur kemampuan model membedakan antara dua kelas (`deposit=yes` dan `deposit=no`).

## **Model Benchmark**

### Alur Pipeline Model Benchmark

Pipeline ini dibangun menggunakan **`ImbPipeline`** agar seluruh tahapan — mulai dari preprocessing, feature selection, resampling, hingga modeling — dapat berjalan secara otomatis serta menghindari *data leakage* selama proses *cross-validation*.  
Setiap model akan melalui tahapan yang sama sehingga hasil perbandingan menjadi adil dan konsisten.


#### 1. Preprocessing
Tahapan preprocessing dibagi menjadi dua bagian: **numerical features** dan **categorical features**.

- Numerical Features
Fitur numerik meliputi: `balance`, `campaign`, `pdays`, dan `age`.

    - Dilakukan **imputasi dengan median**, karena dari hasil EDA diketahui seluruh kolom numerik memiliki **distribusi yang tidak normal dan terdapat outlier ekstrem**, terutama pada kolom `balance` (saldo nasabah) dan `pdays` (jumlah hari sejak kontak terakhir).  
  Nilai median lebih robust dibanding mean, sehingga tidak terpengaruh oleh nilai ekstrem.
  
  - Setelah itu diterapkan **`RobustScaler`**, karena scaler ini menggunakan IQR (interquartile range) dan **tidak sensitif terhadap outlier**.  
  Hal ini penting karena hampir semua variabel numerik menunjukkan adanya outlier saat analisis boxplot, sehingga penggunaan `StandardScaler` atau `MinMaxScaler` akan menyebabkan distorsi skala.

 - Categorical Features
Fitur kategorikal meliputi: `job`, `housing`, `loan`, `contact`, dan `month`.

    - Diterapkan **imputasi dengan `most_frequent`** karena sebagian fitur kategorikal memiliki jumlah kategori terbatas dan distribusi nilai tidak seimbang.  
  Mengisi nilai kosong dengan kategori yang paling sering muncul dinilai paling representatif dan tidak mengubah distribusi data secara signifikan.

    - Semua fitur kategorikal dikonversi menggunakan **`OneHotEncoder (OHE)`**, karena:
    - `job`, `housing`, dan `loan` adalah variabel nominal (tanpa urutan nilai), sehingga OHE cocok untuk menghindari asumsi hierarki antar kategori.  
    - `contact` juga nominal (tipe media kontak: cellular, telephone, unknown).  
    - `month` walaupun tampak berurutan, pada konteks kampanye pemasaran **tidak memiliki hubungan numerik langsung dengan outcome** (misalnya “Maret” tidak selalu lebih besar dari “Januari”), sehingga lebih tepat dianggap nominal dan diencode dengan OHE.
  
Dengan pendekatan ini, seluruh fitur kategorikal diubah menjadi representasi numerik biner tanpa mengasumsikan hubungan ordinal yang tidak valid.


#### 2. Feature Selection
Tahapan **`SelectKBest(score_func=f_classif)`** digunakan untuk memilih fitur-fitur yang memiliki korelasi paling kuat terhadap target (`deposit`).  
Metode ini menggunakan uji statistik **ANOVA F-test**, cocok untuk data dengan target kategorikal biner.  
Pemilihan fitur penting ini membantu:
- Mengurangi kompleksitas model,  
- Meningkatkan generalisasi,  
- Dan mengurangi potensi *overfitting*.

#### 3. Resampling
Data target `deposit` menunjukkan ketidakseimbangan kelas, dengan proporsi nasabah yang melakukan deposit lebih banyak dibanding yang tidak.  
Untuk mengatasi hal ini digunakan **SMOTE (Synthetic Minority Oversampling Technique)**, yang membuat sampel sintetis baru dari kelas minoritas berdasarkan *k-nearest neighbors*.  
Tujuannya agar model tidak bias terhadap kelas mayoritas dan dapat meningkatkan **recall** pada kelas minoritas.

#### 4. Model Benchmark
Model benchmark yang diuji meliputi:
- **Logistic Regression** -> baseline model, interpretatif dan sederhana.  
- **Decision Tree** -> model non-linear dasar untuk melihat pola keputusan.  
- **Random Forest** -> ensemble dari banyak pohon untuk hasil lebih stabil.  
- **KNN** -> metode berbasis jarak untuk pembanding performa sederhana.  
- **XGBoost**, **LightGBM**, dan **CatBoost** -> model boosting modern dengan performa tinggi dan kemampuan generalisasi yang baik pada data kompleks.


#### **Kesimpulan**
Pipeline ini dirancang agar proses modeling berlangsung **terstruktur, konsisten, dan bebas data leakage**.  
Semua model benchmark diproses dengan pipeline yang sama mulai dari imputasi, scaling, encoding, balancing, hingga seleksi fitur sehingga hasil evaluasi dapat merepresentasikan performa murni dari setiap algoritma yang diuji.


```python
# ==== PREPROCESSOR ====
preprocessor = ColumnTransformer(
    transformers=[
        # Balance 
        ('Numerical', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', RobustScaler())
        ]), ['balance','campaign','pdays','age']),

        # Categorical 
        ('Categorical', Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('encoder', OneHotEncoder(drop='first', handle_unknown='ignore'))
        ]), ['job', 'housing', 'loan', 'contact', 'month']),
    ],
    remainder='passthrough'
)

# ==== FEATURE SELECTION ====
select_kbest = SelectKBest(score_func=f_classif, k=10)

# ==== MODELS ====
logreg = LogisticRegression(max_iter=1000, random_state=10)
tree   = DecisionTreeClassifier(random_state=10)
rf     = RandomForestClassifier(random_state=10)
knn    = KNeighborsClassifier()
xgb    = XGBClassifier(random_state=10, eval_metric='logloss', use_label_encoder=False)
lgbm   = LGBMClassifier(random_state=10)
cat    = CatBoostClassifier(random_state=10, verbose=0)

# ==== BASE PIPELINE ====
base_pipeline = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', select_kbest),
    ('resampler', SMOTE(random_state=10)),
    ('classifier', logreg)
])

# ==== GRID ====
param_grid = [
    {'feature_selection__k':[10,20], 'classifier':[logreg]},
    {'feature_selection__k':[10,20], 'classifier':[tree]},
    {'feature_selection__k':[10,20], 'classifier':[rf]},
    {'feature_selection__k':[10,20], 'classifier':[knn]},
    {'feature_selection__k':[10,20], 'classifier':[xgb]},
    {'feature_selection__k':[10,20], 'classifier':[lgbm]},
    {'feature_selection__k':[10,20], 'classifier':[cat]},
]

# ==== GRID SEARCH ====
gscv = GridSearchCV(
    estimator=base_pipeline,
    param_grid=param_grid,
    cv=3,
    n_jobs=-1,
    scoring='f1',
    return_train_score=True
)

import time

#create start time
start_time=time.time()

#fit model
gscv.fit(X_train,y_train)
print("Best Params:", gscv.best_params_)
print("Best CV F1:", gscv.best_score_)

#create end time
end_time=time.time()

fitting_time=end_time-start_time

print(f"fitting time : {fitting_time:.2f} second")

y_pred = gscv.predict(X_test)
print(classification_report(y_test, y_pred, digits=3))
```

```python
# Ambil hasil dari GridSearchCV
results = pd.DataFrame(gscv.cv_results_)

# Ambil kolom penting
model_scores = results[
    ['param_classifier', 
     'param_feature_selection__k', 
     'mean_train_score', 
     'mean_test_score', 
     'std_test_score', 
     'mean_fit_time']
]

# Bersihkan nama model
model_scores['Model'] = model_scores['param_classifier'].apply(lambda x: type(x).__name__)

# Ambil performa terbaik per model
best_models = (
    model_scores.groupby('Model')
    .agg({
        'mean_train_score': 'max',
        'mean_test_score': 'max',
        'std_test_score': 'min',
        'mean_fit_time': 'mean'
    })
    .reset_index()
)

# Hitung selisih train-test untuk indikasi overfitting
best_models['Gap (Train - Test)'] = best_models['mean_train_score'] - best_models['mean_test_score']

# Ganti nama kolom agar lebih rapi
best_models = best_models.rename(columns={
    'mean_train_score': 'F1-score (Train CV Mean)',
    'mean_test_score': 'F1-score (Test CV Mean)',
    'std_test_score': 'F1 Std',
    'mean_fit_time': 'Fit Time (s)'
})

# Urutkan dari yang performanya terbaik
best_models = best_models.sort_values(by='F1-score (Test CV Mean)', ascending=False).reset_index(drop=True)

# Tampilkan hasil ranking
print("Model Benchmark Ranking Berdasarkan F1-score (Train vs Test)")
display(best_models)
```

## **Hyperparameter Tuning**

Setelah melakukan model benchmarking dan memperoleh 5 model terbaik (Logistic Regression, Random Forest, XGBoost, LightGBM, dan CatBoost), langkah berikutnya adalah melakukan **hyperparameter tuning** untuk mengoptimalkan performa masing-masing model.

Tahapan ini menggunakan **`GridSearchCV`** dengan pendekatan pipeline yang terintegrasi agar proses tuning tetap menjaga data leakage dan memastikan semua langkah preprocessing serta balancing dijalankan konsisten di setiap fold cross-validation.

Beberapa alasan utama dilakukan hyperparameter tuning:

1. **Meningkatkan performa model:**  
   Parameter default seringkali belum memberikan hasil optimal. Dengan mencari kombinasi parameter terbaik, model dapat meningkatkan akurasi dan F1-score pada data test.

2. **Mengurangi overfitting:**  
   Parameter seperti `max_depth`, `min_samples_leaf`, atau `C` membantu mengatur kompleksitas model agar tidak hanya fit pada data training saja.

3. **Menentukan efek balancing data:**  
   Pipeline ini menguji dua kondisi resampling dengan dan tanpa **SMOTE** untuk melihat apakah balancing data memberikan pengaruh signifikan terhadap performa model.

4. **Menjaga reproducibility dan fairness antar model:**  
   Semua model dievaluasi dengan preprocessing dan feature selection yang sama (`SelectKBest` dan `RobustScaler`), sehingga perbandingan hasil lebih adil.

Metode tuning ini menggunakan:
- **`GridSearchCV(cv=5)`** → melakukan validasi silang 5 fold untuk memastikan model tidak bias terhadap satu subset data.  
- **`scoring='f1'`** → F1-score dipilih karena dataset target sedikit imbalanced, sehingga metrik ini lebih representatif daripada akurasi.

Output dari tahap ini akan menunjukkan kombinasi parameter terbaik (**Best Params**) dan skor F1 tertinggi (**Best CV F1**) untuk masing-masing model kandidat.


```python
# ==== PREPROCESSOR ====
preprocessor = ColumnTransformer(
    transformers=[
        # Numerical 
        ('Numerical', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', RobustScaler())
        ]), ['balance','campaign','pdays','age']),

        # Categorical
        ('Categorical', Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('encoder', OneHotEncoder(drop='first', handle_unknown='ignore'))
        ]), ['job', 'housing', 'loan', 'contact', 'month']),
    ],
    remainder='passthrough'
)


# ================== FEATURE SELECTION ==================
select_kbest = SelectKBest(score_func=f_classif, k=10)

# ================== MODELS ==================
logreg = LogisticRegression(max_iter=1000, random_state=10)
rf     = RandomForestClassifier(random_state=10)
xgb    = XGBClassifier(random_state=10, eval_metric='logloss', use_label_encoder=False)
lgbm   = LGBMClassifier(random_state=10)
cat    = CatBoostClassifier(random_state=10, verbose=0)

# ================== PIPELINE ==================
base_pipeline = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', select_kbest), 
    ('resampler', 'passthrough'),
    ('classifier', logreg)           
])

# ====== PARAM GRID (Hyperparameter Tuning) ======
param_grid = [
    # Logistic Regression 
    {
     'feature_selection__k': [20, 25],
     'resampler': ['passthrough',SMOTE(random_state=10)],
     'classifier': [logreg],
     'classifier__C': [0.5, 1.0, 2.0],
     'classifier__solver': ['lbfgs'],
     'classifier__penalty': ['l2']
    },

    # Random Forest
    {
     'feature_selection__k': [20, 25],
     'resampler': ['passthrough',SMOTE(random_state=10)],
     'classifier': [rf],
     'classifier__n_estimators': [200, 400],
     'classifier__max_depth': [None, 8, 12],
     'classifier__min_samples_split': [2, 5],
     'classifier__min_samples_leaf': [3, 5],
     'classifier__max_features': ['sqrt']
    },

    # XGBoost
    {
     'feature_selection__k': [20, 25],
     'resampler': ['passthrough',SMOTE(random_state=10)],
     'classifier': [xgb],
     'classifier__n_estimators': [300, 500],
     'classifier__max_depth': [5],
     'classifier__learning_rate': [0.05],
     'classifier__subsample': [0.8, 1.0],
     'classifier__colsample_bytree': [0.8, 1.0],
     'classifier__min_child_weight': [1]
    },

    # LightGBM
    {
     'feature_selection__k': [20, 25],
     'resampler': ['passthrough',SMOTE(random_state=10)],
     'classifier': [lgbm],
     'classifier__n_estimators': [300, 500],
     'classifier__num_leaves': [31, 63],
     'classifier__learning_rate': [0.05],
     'classifier__feature_fraction': [0.8, 1.0],
     'classifier__bagging_fraction': [0.8, 1.0],
     'classifier__bagging_freq': [1],
    },

    # CatBoost
    {
     'feature_selection__k': [20, 25],
     'resampler': ['passthrough',SMOTE(random_state=10)],
     'classifier': [cat],
     'classifier__iterations': [300, 500],
     'classifier__depth': [6],
     'classifier__learning_rate': [0.05],
     'classifier__l2_leaf_reg': [5]
    },
]

# ====== GRID SEARCH  ======
gscv = GridSearchCV(
    estimator=base_pipeline,
    param_grid=param_grid,
    cv=5,                
    n_jobs=-1,
    scoring='f1',
    return_train_score=True
)


import time

#create start time
start_time=time.time()

#fit model
gscv.fit(X_train,y_train)
print("Best Params:", gscv.best_params_)
print("Best CV F1:", gscv.best_score_)

#create end time
end_time=time.time()

fitting_time=end_time-start_time

print(f"fitting time : {fitting_time:.2f} second")
```

## Random Forest Classifier



### 1. Apa itu Random Forest?
**Random Forest** adalah salah satu algoritma *ensemble learning* berbasis pohon keputusan (*Decision Tree*).  
Berbeda dengan satu pohon keputusan yang cenderung mudah overfitting, **Random Forest** membangun banyak pohon (tree) secara acak, kemudian menggabungkan hasil prediksi dari masing-masing pohon untuk menghasilkan keputusan akhir yang lebih stabil dan akurat.

**Model ini bekerja seperti *voting system*, di mana setiap pohon memberikan “suara” untuk hasil prediksi, dan hasil akhir diambil berdasarkan mayoritas suara.**

### 2. Cara Kerja Secara Sederhana

1. **Bootstrapping (Sampling Data)**  
   Dari dataset asli, Random Forest membuat beberapa subset data acak (*bootstrap samples*).  
   Tiap pohon dilatih dengan subset yang berbeda untuk menciptakan variasi.

2. **Random Feature Selection**  
   Saat membangun setiap pohon, hanya sebagian fitur yang dipilih secara acak.  
   Hal ini mencegah semua pohon belajar pola yang sama, sehingga meningkatkan generalisasi model.

3. **Voting (Klasifikasi)**  
   Setelah semua pohon selesai belajar, setiap pohon memberikan hasil prediksi (`ya` atau `tidak deposit`).  
   Prediksi akhir ditentukan berdasarkan **mayoritas suara**.

   Contoh sederhana:
   - Pohon 1 → “Deposit”
   - Pohon 2 → “Tidak Deposit”
   - Pohon 3 → “Deposit”  
   Maka hasil akhir model = “Deposit” (karena 2 dari 3 pohon setuju).

### 3. Kelebihan Random Forest
| No | Kelebihan | Penjelasan |
|:--:|:--|:--|
| 1 | **Akurasi Tinggi** | Menggabungkan banyak pohon membuat prediksi lebih stabil dan akurat. |
| 2 | **Tahan terhadap Overfitting** | Variasi antar-pohon mencegah model terlalu menghafal data latih. |
| 3 | **Dapat Menangani Data Campuran** | Efektif untuk data dengan fitur numerik dan kategorikal sekaligus. |
| 4 | **Memberikan Interpretasi (Feature Importance)** | Dapat mengukur seberapa besar pengaruh tiap fitur terhadap hasil prediksi. |

### 4. Kekurangan Random Forest
| No | Kekurangan | Dampak |
|:--:|:--|:--|
| 1 | **Waktu Latih Lebih Lama** | Karena terdiri dari banyak pohon, komputasinya lebih berat dibanding model tunggal. |
| 2 | **Kurang Interpretatif** | Sulit menjelaskan secara langsung hubungan antar variabel seperti pada regresi. |
| 3 | **Butuh Tuning Hyperparameter** | Agar performa optimal, parameter seperti `n_estimators`, `max_depth`, dan `min_samples_leaf` perlu diuji secara cermat. |

### 5. Mengapa Dipilih sebagai Model Terbaik
Dari hasil *benchmark* dan *hyperparameter tuning*, **Random Forest Classifier** menunjukkan:
- Nilai **F1-score tertinggi (0.85)** dengan selisih *train–test gap* kecil → menandakan model seimbang dan tidak overfit.  
- Memiliki **ROC-AUC 0.83**, menunjukkan kemampuan model dalam membedakan nasabah yang berpotensi deposit dan tidak.  
- Menyediakan **feature importance**, yang sangat berguna untuk analisis faktor-faktor bisnis.

## **Feature Importance (Random Forest)**

```python
# ==== FEATURE IMPORTANCE VISUALIZATION ====
best_model = gscv.best_estimator_.named_steps['classifier']

# Ambil nama fitur setelah feature selection
feature_names = gscv.best_estimator_.named_steps['feature_selection'].get_support(indices=True)
selected_feature_names = np.array(feat_cols)[feature_names]

# Ambil nilai importance sesuai tipe model
if hasattr(best_model, "coef_"):
    coefficients = np.abs(best_model.coef_).flatten()
    model_coef1 = pd.Series(coefficients, index=selected_feature_names)
elif hasattr(best_model, "feature_importances_"):
    importances = best_model.feature_importances_
    model_coef1 = pd.Series(importances, index=selected_feature_names)
else:
    raise ValueError("Model tidak punya coef_ atau feature_importances_")

# Urutkan descending 
model_coef1 = model_coef1.sort_values(ascending=True)

# Plot dengan anotasi
plt.figure(figsize=(15, 8))
ax = model_coef1.plot(kind='barh', title='Feature Importance', color='steelblue')

# Tambahkan anotasi nilai
for i, v in enumerate(model_coef1):
    ax.text(v + 0.002, i, f'{v:.3f}', va='center', fontsize=9, color='black')

plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```
